#Loading the Image Files for Each Class to NumPy Arrays and Then Save as NPY Files

##Meningioma

```python
##################################################
#Populating meningioma array
##################################################
meningioma_no = 708 
meningioma_counter = 0
```

```python
meningioma_ar = np.zeros((meningioma_no, 256, 256, 3), dtype=np.float16)
```

```python
#folder 1
for file in list_of_files_1:
  f = h5py.File(file, 'r')
  label = f['cjdata']['label'][0]
  sq_label = np.squeeze(label).astype(np.uint8)
  if sq_label == 1:
    temp_var = f['cjdata']['image'][:, :]
    temp_var = cv2.resize(temp_var, dsize=(256, 256), interpolation=cv2.INTER_CUBIC)
    min_val = np.amin(temp_var)
    max_val = np.amax(temp_var)
    range_val = (max_val - min_val)
    temp_var = temp_var/range_val
    meningioma_ar[meningioma_counter, :, :, 0] = temp_var
    meningioma_ar[meningioma_counter, :, :, 1] = temp_var
    meningioma_ar[meningioma_counter, :, :, 2] = temp_var
    meningioma_counter += 1
```

```python
print (meningioma_counter)
```
